```python
import os
import pandas as pd
from sentence_transformers import SentenceTransformer
import numpy as np
```

```python
base_dir = "../wheat_all/done/"
files = os.listdir(base_dir)
```

```python
env_df = pd.DataFrame()
for folder in files:
    file_path = os.path.join(base_dir, folder, 'EnvData.xls')
    df = pd.read_csv(file_path, delimiter='\t', encoding = "ISO-8859-1")
    df.drop(['Trial name', 'Occ', 'Country', 'Loc_desc', 'Cycle', 'Trait No', 'Unit'], axis=1, inplace=True)
    df['trial'] = folder
    env_df = pd.concat([env_df,df])
```

```python
env_df.dropna(inplace=True)
```

```python
rows_to_drop = ["Cooperator's  E-mail", 'COOPERATOR_NOTE_TAKER', 
                'DECIMAL_CODE_STAGE', 'EMERGENCE', 'EMERGENCE_DATE', 
               'GPS Altitude', 'GPS Latitude (Degrees)',
               'GPS Latitude (Minutes)', 'GPS Latitude (N or S)',
               'GPS Longitude ( E or W)', 'GPS Longitude (Degress)',
               'GPS Longitude (Minutes)', 'HAIL_DAMAGE', 'HARVEST_FINISHING_DATE',
               'HARVEST_FINISHING_DATE_TEXT', 'HARVEST_STARTING_DATE',
               'HERBICIDE_PRODUCT(S)', 'LENGTH_OF_ROWS_HARVESTED', 'LENGTH_OF_ROWS_SOWN',
               'NO_OF_ROWS_HARVESTED', 'NO_OF_ROWS_SOWN',
               'NUMBER_POST_SOWING_IRRIGATIONS', 'OTHER_CHEMICAL_PRODUCT(S)',
               'OTHER_COMMENTS_AND_OBSERVATIONS', 'OTHER_ELEMENT(S)_1',
               'OTHER_ELEMENT(S)_2', 'PESTICIDE_PRODUCT(S)'
               'SOWING_DATE', 'SPACE_BTN_ROWS_SOWN',
                'WEATHER_COMMENTS', 'EMERGENCE_DATE_TEXT',
               'FUNGICIDE_PRODUCT(S)', 'AREA_HARVESTED_BED_PLOT_M2', 
               'SOWING_DATE', 'TOTAL_PRECIPIT_IN_12_MONTHS',
                'ESTIMATE_TOTAL_PRECIPIT_IN_12_MONTHS',
               'PRECIPITATION_ON_CROP', 
                'SPACE_BTN_ROWS_HARVESTED']
```

```python
for row in rows_to_drop:
    env_df = env_df[env_df['Trait name'] != row]
```

```python
env_df['Trait name'].unique()
```

```python
locations = env_df['Loc_no'].unique()
trials = env_df['trial'].unique()
```

```python
model = SentenceTransformer('recobo/agri-sentence-transformer', device='cpu')
representation_df = pd.DataFrame()
```

```python
for loc in locations:

    for trial in trials:
        
        x = env_df[(env_df['Loc_no'] == loc) & (env_df['trial'] == trial)]
        data = []
        representations = 0
        
        if len(x) > 0:
            text = ''
            for ind, row in x.iterrows():
                text = text + row['Trait name'].lower() 
                text = text.replace('_', ' ')
                text = text + ' ' + row['Value'].lower() + ' '
                # data.append(text)
                
            embedding = model.encode(text)
            avg_embedding = np.average(embedding, axis=0)
           
            
            dic = {'trial': [trial], 'Loc_no': [loc], 'embedding': [avg_embedding.tolist()], 'text': text}
            dic = pd.DataFrame(dic)
            representation_df = pd.concat([representation_df, pd.DataFrame(dic)], ignore_index = True, axis=0)
            
            
           
```
